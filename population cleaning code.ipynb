```python

population = pd.read_csv('SD1970_population.csv',thousands = ',')

#A duplicate dataframe
populationCopy = population

populationCopy = populationCopy.replace('...',np.nan)


"""
This function is designed specifically for the population dataset.

It takes in:
1. Name of the leftmost column you want to merge
2. Name of the rightmost column you want to merge
3. New name of the merged column.

It merges everything in place.

"""
def ageMerge(left_name, right_name, new_name):
    i = populationCopy.columns.get_loc(left_name)
    j = populationCopy.columns.get_loc(right_name) + 1
    populationCopy[left_name] = populationCopy.iloc[:,i:j].astype(float).sum(axis=1)
    populationCopy.rename(columns = {left_name: new_name},inplace = True)
    populationCopy.drop(populationCopy.iloc[:, i+1:j], inplace = True, axis = 1)

ageMerge('Male persons age 5 years', 'Male persons age 22-24 years', 'Male persons 5-24 years')

ageMerge('Male persons age 25-34 years', 'Male persons age 35-44 years', 'Male persons 25-44 years')

ageMerge('Male persons age 45-54 years', 'Male persons age 62-64 years', 'Male persons 45-64 years')



populationCopy
```
