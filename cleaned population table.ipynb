```python
population = pd.read_csv('SD1970_population.csv',thousands = ',')

#A duplicate dataframe
populationCopy = population

populationCopy = populationCopy.replace('...',np.nan)


"""
This function is designed specifically for the population dataset.

It takes in:
1. Name of the leftmost column you want to merge
2. Name of the rightmost column you want to merge
3. New name of the merged column.

It merges everything in place.

"""
def ageMerge(left_name, right_name, new_name):
    i = populationCopy.columns.get_loc(left_name)
    j = populationCopy.columns.get_loc(right_name) + 1
    populationCopy[left_name] = populationCopy.iloc[:,i:j].astype(float).sum(axis=1)
    populationCopy.rename(columns={left_name: new_name},inplace=True)
    populationCopy.drop(populationCopy.iloc[:, i+1:j], inplace = True, axis = 1)


ageMerge('Male persons age under 5 years','Male persons age 75 and over','Total Male Persons')
ageMerge('Female persons age under 5 years','Female persons age 75 and over','Total Female Persons')

ageMerge('Black male persons age under 5 years','Black male persons age 65 and over','Total Black Male Persons')
ageMerge('Black female persons age under 5 years','Black female persons age 65 and over','Total Black Female Persons')

ageMerge('Other nonwhite races male age under 5 years','Other nonwhite races male age 65 and over','Total Nonwhite Male Persons')
ageMerge('Other nonwhite races female age under 5 years','Other nonwhite races female age 65 and over','Total Nonwhite Female Persons')



populationCopy=populationCopy.drop(populationCopy.loc[:, 'Male persons now married (excludes separated)':'Family with female head members under 18 and 65 and over'].columns, axis = 1)

populationCopy
```
