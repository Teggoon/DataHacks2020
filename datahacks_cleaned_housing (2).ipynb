```python
#[x[1:] for x in housing2.owner_occupied_aggregate_values]
housing2 = housing2.replace('...', np.NaN)
housing2
```

```python
for column in housing2.columns:
    #col_name = column
    if housing2[column].dtype == str:
        housing2[column] = pd.to_numeric(housing2[column])
        
columns_with_money = ["owner_occupied_aggregate_values","owner_occupied_average_values","renter_occupied_aggregate_rents"];

for name in columns_with_money:
    housing2[name] = housing2[name].str.replace(',', '')
    housing2[name] = housing2[name].str.replace('$', '')
    housing2[name] = housing2[name].str.replace('.00', '')

housing2
```
